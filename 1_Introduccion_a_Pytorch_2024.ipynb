## 游빑 Operaciones b치sicas con tensores

Similar a lo que sucede en `numpy`, podemos realizar muchas operaciones utilizando tensores como slicing, transposici칩n, multiplicaci칩n de matrices, entre otras.


### Suma

````python
# Suma de tensores
print('Tensor identidad:')
print(tensor_eye)
print('Tensor de unos:')
print(tensor_ones )
print('\n')

print('Tensor suma:')
tensor_suma = tensor_eye + tensor_ones
print(tensor_suma)
````

### Resta

````python
# Resta de tensores
tensor_resta = tensor_ones - tensor_eye
print(tensor_resta)
````

### Producto de elementos de tensores (element-wise product)

````python
# Producto de elementos de tensores (element-wise product)
tensor_producto = tensor_suma * tensor_resta
print(f'{tensor_suma} \n* {tensor_resta} \n= {tensor_producto}')
````

### Producto de matrices

*  https://en.wikipedia.org/wiki/Hadamard_product_(matrices)
*  https://pytorch.org/docs/stable/generated/torch.matmul.html



````python
# Producto de matrices, tensor suma con tensor resta (producto punto)
tensor_prodmat = torch.matmul(tensor_suma, tensor_resta)
print(f'{tensor_suma} \n* {tensor_resta} \n= {tensor_prodmat}')
````

## Calculando derivadas con pytorch

En general, las redes neuronales artificiales se entrenan usando el algoritmo del **descenso de gradiente**. Por lo tanto necesitamos calcular las derivadas de la funci칩n de costo para todos los par치metros de la red. La **Autodiferenciaci칩n** es una t칠cnica que permite calcular autom치ticamente las derivadas de una funci칩n respecto a sus entradas. PyTorch construye autom치ticamente el grafo de c칩mputo durante la ejecuci칩n de operaciones. Este grafo es una representaci칩n de c칩mo los tensores est치n conectados a trav칠s de operaciones matem치ticas.

PyTorch tiene incorporado un sistema de diferenciaci칩n autom치tica denominado autograd. Para poder derivar una funci칩n en pytorch:



1. Se necesita que su entrada sean tensores con el atributo `requires_grad=True`.
2. Luego llamamos la funci칩n `backward()` de la funci칩n
3. El resultado queda guardado en el atributo `grad` de la entrada (nodo hoja)

````python
x = torch.linspace(0, 10, steps=1000, requires_grad=True)
y = 5*x - 20
y.backward(torch.ones_like(x))
````

````python
fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(x.detach().numpy(), y.detach().numpy(), label='y')
ax.plot(x.detach().numpy(), x.grad.detach().numpy(), label='dy/dx') #detach genera una copia del tensor que no tiene informacion de los gradientes
plt.legend();
````

````python
x1 = torch.linspace(0, 10, steps=1000, requires_grad=True)
y1 = torch.sin(2.0*np.pi*x1)*torch.exp(-(x1-5).pow(2)/3)
y1.backward(torch.ones_like(x1))

fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(x1.detach().numpy(), y1.detach().numpy(), label='y1')
ax.plot(x1.detach().numpy(), x1.grad.detach().numpy(), label='dy1/dx1')
plt.legend();
````

Comparado con la derivada calculada 라 mano:

````python
dydx = 2*torch.pi*torch.cos(2.0*np.pi*x1)*torch.exp(-(x1-5).pow(2)/3) - 2/3*(x1-5)*torch.sin(2.0*np.pi*x1)*torch.exp(-(x1-5).pow(2)/3)
torch.allclose(x1.grad.detach(), dydx)
````

Un poco m치s de detalle: En PyTorch, el m칠todo `detach()` se utiliza para crear un nuevo tensor que comparte el mismo almacenamiento de datos que el tensor original, pero que no forma parte del grafo de c칩mputo para el c치lculo de gradientes. Esto significa que el tensor resultante de `detach()` no tendr치 gradientes calculados durante la retropropagaci칩n (`.backward()`), y no se actualizar치n los gradientes de los tensores originales a partir de 칠l.

### Grafo de computo

Cuando concatenamos operacionesm PyTorch construye internamente un 랂rafo de c칩mputo렢nCaptura de pantalla 2024-09-14 a la(s) 5.14.28슳.맔..png

El m칠todo `backward()` calcula los gradientes y los almacena en los nodo hoja que tengan `requires_grad=True`.

Por ejemplo,

`y.backward` : Guarda dy/dx en x.grad

`z.backward` : Guarda dz/dx en x.grad

````python
x = torch.linspace(0, 10, steps=1000, requires_grad=True)
print(x.grad_fn)
````

````python
z = torch.sin(2*x)
print(z.grad_fn)
````

````python
y = z.pow(2)/2
print(y.grad_fn)
````
