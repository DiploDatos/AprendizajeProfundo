## Tensores

En Deep Learning, los tensores son una estructura de datos fundamental (muy similares a los arreglos y matrices con los que venimos trabajando) con los cuales es posible realizar operaciones matemáticas de manera eficiente usando **grandes conjuntos de datos**. Un tensor se puede representar como una matriz, un vector, un escalar o un arreglo de varias dimensiones.

Desde un punto de vista visual, es posible pensar un tensor como un arreglo simple que contiene escalares u otros arreglos. En Pytorch, ***un tensor es una estructura muy similar a un `ndarray` de la libreria ```numpy``` con la diferencia que son capaces de correr en GPU***, lo cual aumenta considerablemente el tiempo computacional.

## ☕ Atributos de los Tensores

Un tensor de Pytorch posee tres atributos importante:

- `shape`: El tamaño del tensor (cuántos elementos hay en cada dimensión del tensor).
- `dtype`: El tipo de datos.
- `device`: El dispositivo donde el tensor es almacenado (puede ser cpu o gpu).

Adicionalmente, existe un método llamado `dim` que indica cuántas dimensiones tiene el tensor (cuantos ejes)

**Nota:** Observar la diferencia entre `shape` y `dim`.

## 🧮 Operaciones básicas con tensores

Similar a lo que sucede en `numpy`, podemos realizar muchas operaciones utilizando tensores como slicing, transposición, multiplicación de matrices, entre otras.


### Suma

````python
# Suma de tensores
print('Tensor identidad:')
print(tensor_eye)
print('Tensor de unos:')
print(tensor_ones )
print('\n')

print('Tensor suma:')
tensor_suma = tensor_eye + tensor_ones
print(tensor_suma)
````

### Resta

````python
# Resta de tensores
tensor_resta = tensor_ones - tensor_eye
print(tensor_resta)
````

### Producto de elementos de tensores (element-wise product)

````python
# Producto de elementos de tensores (element-wise product)
tensor_producto = tensor_suma * tensor_resta
print(f'{tensor_suma} \n* {tensor_resta} \n= {tensor_producto}')
````

### Producto de matrices

*  https://en.wikipedia.org/wiki/Hadamard_product_(matrices)
*  https://pytorch.org/docs/stable/generated/torch.matmul.html



````python
# Producto de matrices, tensor suma con tensor resta (producto punto)
tensor_prodmat = torch.matmul(tensor_suma, tensor_resta)
print(f'{tensor_suma} \n* {tensor_resta} \n= {tensor_prodmat}')
````

## Calculando derivadas con pytorch

En general, las redes neuronales artificiales se entrenan usando el algoritmo del **descenso de gradiente**. Por lo tanto necesitamos calcular las derivadas de la función de costo para todos los parámetros de la red. La **Autodiferenciación** es una técnica que permite calcular automáticamente las derivadas de una función respecto a sus entradas. PyTorch construye automáticamente el grafo de cómputo durante la ejecución de operaciones. Este grafo es una representación de cómo los tensores están conectados a través de operaciones matemáticas.

PyTorch tiene incorporado un sistema de diferenciación automática denominado autograd. Para poder derivar una función en pytorch:



1. Se necesita que su entrada sean tensores con el atributo `requires_grad=True`.
2. Luego llamamos la función `backward()` de la función
3. El resultado queda guardado en el atributo `grad` de la entrada (nodo hoja)

````python
x = torch.linspace(0, 10, steps=1000, requires_grad=True)
y = 5*x - 20
y.backward(torch.ones_like(x))
````

````python
fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(x.detach().numpy(), y.detach().numpy(), label='y')
ax.plot(x.detach().numpy(), x.grad.detach().numpy(), label='dy/dx') #detach genera una copia del tensor que no tiene informacion de los gradientes
plt.legend();
````

````python
x1 = torch.linspace(0, 10, steps=1000, requires_grad=True)
y1 = torch.sin(2.0*np.pi*x1)*torch.exp(-(x1-5).pow(2)/3)
y1.backward(torch.ones_like(x1))

fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(x1.detach().numpy(), y1.detach().numpy(), label='y1')
ax.plot(x1.detach().numpy(), x1.grad.detach().numpy(), label='dy1/dx1')
plt.legend();
````

Comparado con la derivada calculada “a mano”:

````python
dydx = 2*torch.pi*torch.cos(2.0*np.pi*x1)*torch.exp(-(x1-5).pow(2)/3) - 2/3*(x1-5)*torch.sin(2.0*np.pi*x1)*torch.exp(-(x1-5).pow(2)/3)
torch.allclose(x1.grad.detach(), dydx)
````

Un poco más de detalle: En PyTorch, el método `detach()` se utiliza para crear un nuevo tensor que comparte el mismo almacenamiento de datos que el tensor original, pero que no forma parte del grafo de cómputo para el cálculo de gradientes. Esto significa que el tensor resultante de `detach()` no tendrá gradientes calculados durante la retropropagación (`.backward()`), y no se actualizarán los gradientes de los tensores originales a partir de él.

### Grafo de computo

Cuando concatenamos operacionesm PyTorch construye internamente un “grafo de cómputo”
Captura de pantalla 2024-09-14 a la(s) 5.14.28 p. m..png

El método `backward()` calcula los gradientes y los almacena en los nodo hoja que tengan `requires_grad=True`.

Por ejemplo,

`y.backward` : Guarda dy/dx en x.grad

`z.backward` : Guarda dz/dx en x.grad

````python
x = torch.linspace(0, 10, steps=1000, requires_grad=True)
print(x.grad_fn)
````

````python
z = torch.sin(2*x)
print(z.grad_fn)
````

````python
y = z.pow(2)/2
print(y.grad_fn)
````

````python
fig, ax = plt.subplots(figsize=(6, 3), tight_layout=True)
ax.plot(x.detach().numpy(), z.detach().numpy(), label='z')
ax.plot(x.detach().numpy(), y.detach().numpy(), label='y')
# Derivada dy/dx
y.backward(torch.ones_like(x), retain_graph=True)
ax.plot(x.detach().numpy(), x.grad.detach().numpy(), label='dy/dx')
# Borro el resultado en x.grad
x.grad = None
# Derivada dz/dx
z.backward(torch.ones_like(x))
ax.plot(x.detach().numpy(), x.grad.detach().numpy(), label='dz/dx')
plt.legend();
````
