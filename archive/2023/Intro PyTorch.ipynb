## Operaciones básicas con tensores 🧮

```python
# Suma de tensores
tensor_suma = tensor_eye + tensor_ones
print(tensor_suma)
```

```python
# Resta de tensores
tensor_resta = tensor_ones - tensor_eye
print(tensor_resta)
```

```python
# Producto de elementos de tensores (element-wise product)
tensor_producto = tensor_suma * tensor_resta
print(f'{tensor_suma} \n* {tensor_resta} \n= {tensor_producto}')
```

https://en.wikipedia.org/wiki/Hadamard_product_(matrices)

```python
# Producto de matrices, tensor suma con tensor resta (producto punto)
tensor_prodmat = torch.matmul(tensor_suma, tensor_resta)
print(f'{tensor_suma} \n* {tensor_resta} \n= {tensor_prodmat}')
```

## Operación a Corazón Abierto de una Red Neuronal  👩‍⚕️💖🕵️‍♂️

```python
# Veamos cómo serían las entradas y salidas, pesos y bias, de una capa
input_tensor = torch.Tensor([[8, 5, 6, 2, 4, 5, 7]])
n = 4
# Capa Lineal
linear_layer = nn.Linear(in_features=7, out_features=n)
output = linear_layer(input_tensor)
print(output)
```

```python
for name, param in linear_layer.named_parameters():
  print(f'{name}: \n\t{param.data}\n\tDimensiones: {param.data.shape}')
```

```python
# Detalle del cálculo hecho por la capa
torch.matmul(input_tensor, linear_layer.weight.T) + linear_layer.bias
```

```python
# Veamos cómo serían las entradas y salidas de las capas de una pequeña red
input_tensor = torch.Tensor([[8, 5, 6, 2, 4, 5, 7]])
n = 10
# Red neuronal con 2 capas (entrada y salida)
model = nn.Sequential(nn.Linear(in_features=7, out_features=n),
                      nn.Linear(n, 1)
                     )

output = model(input_tensor)
print(output)
```

```python
for name, param in model.named_parameters():
  # if param.requires_grad:
    print(f'{name}: \n\t{param.data}\n\tDimensiones: {param.data.shape}')
```

```python
# Pesos y sesgos de la capa de entrada (capa 0)
l1_weights = model[0].weight
l1_bias = model[0].bias
l1_output = torch.matmul(input_tensor, l1_weights.T) + l1_bias
l1_output
```

```python
# Función para obtener los pesos y sesgos de una capa
def get_weights_and_bias(layer_number, model):
  weights = model[layer_number].weight
  bias = model[layer_number].bias
  return weights, bias

# Pesos y sesgos de la capa de salida (capa 1)
layer_number = 1
l2_weights, l2_bias = get_weights_and_bias(layer_number, model)

# Salida de la capa 1
l2_output = torch.matmul(l1_output, l2_weights.T) + l2_bias
l2_output
```

## Función Sigmoide  📈

```python
# Instanciar función sigmoide y aplicar a la salida de la red
# Create a sigmoid function and apply it on the score tensor
sigmoid = nn.Sigmoid()
probability = sigmoid(l2_output)
print(probability)
```

```python
# La red neuronal con función sigmoide como salida
model = nn.Sequential(
  nn.Linear(input_tensor.shape[1], 1),
  nn.Sigmoid()
)

output = model(input_tensor)
print(output)
```

## Función Softmax 📈

```python
# Simulando más de 2 salidas (clasificación múltiple)
outputs = torch.tensor([[-3.7, -2.5, -0.2,  0.8, 1.8, 3.2]])

# Instanciar función sigmoide y aplicar a distintos valores
softmax = nn.Softmax(dim=-1)
probabilities = softmax(outputs)
print(probabilities)
```

```python
# La red neuronal con función sigmoide como salida
model = nn.Sequential(
  nn.Linear(input_tensor.shape[1], 3),
  nn.Softmax(dim=-1)
)

output = model(input_tensor)
print(output)
```
